```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
database = pd.read_excel("linear_regression_dummy_data.xlsx")
database.head(3)
```

```python
database.shape
```

```python
x = database.iloc[:, :-1].values
y = database["Y"].values
x
```

```python
x = x.reshape(-1, 1)
x
```

```python
database.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
llr = LinearRegression()
llr.fit(x_train, y_train)
```

```python
llr.score(x_test, y_test)*100
```

```python
predict_data = llr.predict(x_test)
predict_data
```

```python
r2 = r2_score(y_test, predict_data)*100
print("R2 Score: ", r2)
```

<h1 style="color:green" align="center">plot the value of X_predict</h1>

```python
plt.scatter(x_test,y_test, color="red", label="Original Data")
plt.plot(x_test, predict_data, color="blue", label="Predicted Data")
plt.show()
```

```python
predict: float = llr.predict([[19.87156815341724]])
print("Predicted salary: ", predict)
```
