```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
database = pd.read_excel("polynomial_regression.xlsx")
database.head(5)
```

```python
x = database[["Experience (Years)"]]
y = database["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)

# pf.fit(x.values.reshape(-1, 1))
# pf.transform(x.values.reshape(-1, 1))
```

```python
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
plt.scatter(x_test[:, 1], y_test, color="red", label="Original Data")
plt.plot(x_test[:, 1], lr.predict(x_test), color="blue", label="Predicted Data")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
