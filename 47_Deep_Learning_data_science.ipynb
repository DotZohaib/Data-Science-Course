```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
database = pd.read_excel("SVM_linear_separable_data.xlsx")
database.head()
```

```python
x = database.iloc[:, :-1]
y = database["Score"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
sns.scatterplot(data=database, x="Data1", y="Data2", hue="Score")
plt.show()
```

```python
from sklearn.linear_model import Perceptron
```

```python
pt = Perceptron()
pt.fit(x_train, y_train)
```

```python
pt.score(x_test, y_test)*100, pt.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pt)
plt.show()
```
