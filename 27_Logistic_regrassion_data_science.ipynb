```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
database = pd.read_excel("logistic_regression_data.xlsx")
database.head(3)
```

```python
x = database.iloc[:, :-1]
y = database["Purchased"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=5)
pf.fit(x)
x = pd.DataFrame(pf.transform(x))
```

```python
x.shape
```

```python
sns.scatterplot(x = 'Age', y = 'Salary', data = database, hue='Purchased')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

<h1>Logistic Regressionn With OVR</h1>


```python
lr1 = LogisticRegression(multi_class='ovr')
lr1.fit(x_train, y_train)
```

```python
lr1.score(x_test, y_test)*100
```

<h1>Logistic Regrassion with multinomial</h1>

```python
lr2 = LogisticRegression(multi_class='multinomial')
lr2.fit(x_train, y_train)
```

```python
lr2.score(x_test, y_test)*100
```
