```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
database = pd.read_excel("KMeans.xlsx")
database.head(5)
```

```python
sns.pairplot(data=database)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20, 10))
sch.dendrogram(sch.linkage(database, method="single", metric="euclidean"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage="single")
database["Predict"] = ac.fit_predict(database)
```

```python
database
```

```python
sns.pairplot(data=database, hue="Predict")
plt.show()
```
