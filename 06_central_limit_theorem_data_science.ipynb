```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = [np.random.randint(10,100) for i in range(10000)]
pop_table = pd.DataFrame({"Population":pop_data})
```

```python
sns.kdeplot(x = "Population", data=pop_table )
plt.show()
```

```python
sample_mean = []
for no_sample in range(60):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
sample_data_table = pd.DataFrame({"Sample Mean":sample_mean})
```

```python
sns.kdeplot(x = "Sample Mean", data=sample_data_table )
plt.show()
```
