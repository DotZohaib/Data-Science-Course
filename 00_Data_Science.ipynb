# What Do Data Scientists Do?

## Overview
Data science is the study of large quantities of data, which can reveal insights that help organizations make strategic choices and drive informed decision-making.

---

## Paths to Data Science

There are many paths to a career in data science. While most involve **mathematics**, **programming**, and a natural **curiosity about data**, there is no single required trajectory. Different backgrounds—statistics, computer science, business, domain expertise—can all lead to a successful data science career.

---

## Essential Traits of Data Scientists

Successful data scientists share three critical attributes:

- **Curiosity:** The drive to ask questions and explore data for hidden patterns and insights
- **Judgment:** The ability to evaluate data quality, assess validity of findings, and make informed decisions about analysis approaches
- **Argumentation:** The skill to defend findings, challenge assumptions, and communicate conclusions persuasively to stakeholders

---

## Career Outlook and Compensation

Knowledgeable data scientists are in **high demand** in today's job market. The field offers:
- **Competitive salaries** for skilled workers
- Strong job growth and career advancement opportunities
- Opportunities across diverse industries and sectors

---

## The Data Scientist's Work

### Varied Daily Activities
The typical work day for a data scientist varies significantly depending on the type of project they are working on:
- **Exploratory projects:** analyzing new datasets to discover patterns
- **Predictive modeling:** building and optimizing algorithms
- **Data pipeline development:** creating systems to process and manage data
- **Stakeholder communication:** presenting findings and recommendations

### Algorithms and Insights
Many **algorithms** are used to extract insights from data. The choice of algorithm depends on:
- The type of data available
- The business question being asked
- The desired outcome (classification, prediction, clustering, etc.)

---

## Key Terminology

Essential data science concepts covered in this lesson:

| Term | Definition |
|------|-----------|
| **Outliers** | Data points that deviate significantly from the norm and may require special handling |
| **Model** | A mathematical representation of relationships in data, used for prediction or classification |
| **Algorithms** | Step-by-step procedures used to solve problems and extract patterns from data |
| **JSON** | A lightweight data format for storing and exchanging structured data |
| **XML** | A markup language for encoding data in a human-readable format |
| **CSV** | Comma-separated values format for storing tabular data |
| **Regression** | A statistical method for modeling relationships between variables |
