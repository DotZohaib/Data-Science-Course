```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
database = pd.read_excel("dummy_users.xlsx")
database.head(5)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# ordinal encoding
od = OrdinalEncoder()
database["Role_Label"] = od.fit_transform(database[["role"]])
```

```python
database
```

```python
sns.heatmap(database[["Role_Label"]])
plt.show()
```

```python
# map function used 

database["Map_Role_function"] = database["role"].map({"Developer": 1, "Designer": 2, "Manager": 3, "QA": 4, "HR": 5, "IT": 6})
database
```
