```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
database = pd.read_excel("dummy_users.xlsx")
database.head(5)
```

```python
database.fillna(database.mode().iloc[0], inplace=True)
```

```python
database
```

```python
# replace data type

database["id"] = database["id"].astype("int64")
database["name"] = database["name"].astype("string")
database["role"] = database["role"].astype("string")
database["salary"] = database["salary"].astype("int64")
database["age"] = database["age"].astype("int64")
database["gender"] = database["gender"].astype("string")
database["alive"] = database["alive"].map({'Yes': True, 'No': False}).astype("boolean")
database["department"] = database["department"].astype("string")
database["alone"] = database["alone"].map({'Yes': True, 'No': False}).astype("boolean")
database["phone_number"] = database["phone_number"].astype("string")
database["city"] = database["city"].astype("string")
database["state"] = database["state"].astype("string")
database["single"] = database["single"].map({'Yes': True, 'No': False}).astype("boolean")

```
