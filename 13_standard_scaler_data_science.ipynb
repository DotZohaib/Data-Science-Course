```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
database = pd.read_excel("dummy_users.xlsx")
database.head(5)
```

```python
database.isnull().sum()
```

```python
database.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(database[['salary']])
```

```python
database["salary_ss"] = pd.DataFrame(ss.transform(database[['salary']]), columns=["x"])
```

```python
database
```

```python
database.describe()
```

```python

plt.figure(figsize=(12, 5))

# First subplot
plt.subplot(1, 2, 1)
plt.title("Before Standardization", fontsize=12, pad=10)
sns.histplot(data=database, x="salary", kde=True)
plt.xlabel("Salary")
plt.ylabel("Count")

# Second subplot
plt.subplot(1, 2, 2)
plt.title("After Standardization", fontsize=12, pad=10)
sns.histplot(data=database, x="salary_ss", kde=True)
plt.xlabel("Standardized Salary")
plt.ylabel("Count")

# Add spacing between subplots and display
plt.tight_layout()
plt.show()
```
