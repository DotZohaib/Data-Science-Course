```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
database = pd.read_excel("logistic_regression_data.xlsx")
database.head(5)
```

```python
x = database.iloc[:, :-1]
y = database["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(x)
x = pd.DataFrame(ss.transform(x), columns=x.columns)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc = DecisionTreeClassifier(max_depth=6)
dc.fit(x_train, y_train)
```

```python
dc.score(x_test, y_test)*100,  dc.score(x_train, y_train)*100,
```

```python
# post pruning 
for i in range(1,20):
    dc1 = DecisionTreeClassifier(max_depth=i)
    dc1.fit(x_train, y_train)
    print(i, dc1.score(x_train, y_train), dc1.score(x_test, y_test))
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=dc)
plt.show()
```
